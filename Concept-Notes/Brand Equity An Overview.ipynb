### What is a brand?
* a brand is an organization's promise to customer to deliver what the brand stands for, functionally and emotionally
* it's a journey, relationship, based on perceptions and experiences that a customer has every time hr or she connects to the brand
* a successful brand is a promise made to the consumer about the good or service that they are purchasing
* brand's promise is important when the customer cannot verify the important attributes before purchasing
    * hidden values - e.g. product origin, quality, long term experience

### Brand Associations
* brand - collective perceptions of an organization's key constituents (customers, suppliers, investors, employees)
* brand is how those constituents experience your products and services and the communications around them
* goal of marketer
    * build and maintain a brand through marketing activities that will move target customers along a continuum of commitment

### Brand Vision
* the way that a brand wants to present itself, what it wants to be
* articulated description of aspirational image for the brand
    * what you want the brand to stand for in the eyes of customers and employees
* when the vision clicks, it will reflect and support biz strategy, differentiate the brand from competitors, resonate with customers, energize employees and partners, precipitate a gush of brand-building ideas

### Brand Loyalty
* from I am aware (prescence) -> I like what it offers (relevance) -> It delivers (performance) -> it offers something more (advantage) -> nothing else beats it (bonding)
* loyalty means customers search for it, wants it and you can ask a price premium

### Net Promoter Score
* strong viral potential - net promtoer score
* customers can be divided into 3:
    * promoters
    * passives
    * detractors
* score = promoters (9-10) - detractors (1-6)

* Centra is an online employee training program
* Had field sales that worked well, now adding telesales
* 2 ideas on how to reduce friction between the 2 types of sales
    * divide by customer market position
    * divide by role (tele - prospecting, field - closing)

### Market
* 70% of corporate training expenditure is in classroom learning
    * other 30% is on training material, e.g. books
* The scope of teaching was broad. Firms trained their employees to improve job skills, culture and morale, they briefed sales forces and customers about new products, and they ran conferences, seminars and workshops to market to prospects, to educate customers and vendors, and to motivate distributors.
* segemtns of the market
    * content provider
    * delivery platform (Centra has 50% market)
    * learning management system

### Centra history
* 3 products
    * eMeeting
    * Conference
    * Symposium (a superset of both of the above)
* avaiable in product or ASP packages 
    * product menas self-host (5 days on boarding)
    * ASP is 3rd party hosted (3 days)
* pricing
    * service for ASP
    * license for product

### Go To Market
* want market penetration, since this is a highly crowded space and only 2-3 will survive
* focused on serving high end, Global 2000 corps
* 30k to 150k ea, 440 install base
* value prop: fees can be covered in less than 1 year
* pitch - need to connect value prop to one of the top 5 priorities of a CEO
* distribution
    * field sales
    * tele
    * alliance partners
        * type 1 - selling infrastrucutre - e.g. microsoft
        * type 2 - system integrators - PWC, Accenture
        * type 3 - LMS providers 
* sales process
    * field sales - telemarketing qualifies from a lead, then goes to field
        * 4 month sales cycle
    * tele sales
        * mostly focused on the eMeeting product
    * enterprise sales

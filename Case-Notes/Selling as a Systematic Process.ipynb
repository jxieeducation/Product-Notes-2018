### divide selling process into a series of distinct steps
* generating attn and interest
* identifying requirements
* gaining conditional commitment
* developing the sale
* closing
* ** how you do these depends on the product, customer sophistication, overall strategy **

### generating attention and interest
* develop personal rapport and credibility and create interest in the product
* establish rapport - e.g. personal discussion
    * if little common in background, ask open ended questions (e.g. tell me your background, what is your strategy?)
* if you can't establish rapport, then move onto selling, so that the convo has a point
* the best method for generating interest is by **citing how the product helped someone else**
    * when not available, cite competitive adv
    * ppl r only interested, if it helps meeting their needs
* to confirm success, ask close-ended question
    * Do you think this is applicable in your biz?
    * Are you interested in discussing further?


### Identifying Requirements
* identify needs that you must satisfy so that later the needs can be mapped to specific product benefits
* open ended questions to flush out requirements
    * tell me about ur biz concerns and issues
    * what is the ideal solution to your problem?
* need to be asked with customer's concerns and internal politics in mind
* if reluctance to talk about biz problem, use a funneling method
    * ask in sequence: key goals, quantative objectives, strategies, specific reqs

### gaining conditional commitment
* develop an understanding of decision process, to establish implicit contract
* confirm a true understnading of customer's reqs
* then ask if how the decision will be made 
* asking for commitment prevents you from wasting time for someone with no interest in purchasing your product

### Developing the sale
* develop a product or solution and gain commitment from key decision makers that it satisfies their needs
* gaining a commitment
    * state the perceived customer need
    * state the product benefit that satisfies the need
    * explain how specific feature meets the needs
    * ask a close ended question that the customer also thinks that it fits the bill

### Closing
* explicitly asking for the order
* several techniques
    * direct close
    * subordinate question close - ask a minor question that implies the major decision (e.g. what color do you like?)
    * window of opportunity close - ask for order based on benefit of having it before an important event
    * puppy dog close - ask for a trial order with no obligations
    * multiple choice close - do you want a 20-inch or 25inch?
* handle objections after asking the questions
    * listen for the real objection
    * emphasize with the customer
    * address the concern
        * direct answer
        * outweigh objection
        * minimize objection
        * reverse objection - turn objection into reason to buy - e.g. thiefs avoid heavier computer
    * question to make sure that the objection is handled
    * reattempt to close
* tips:
    * never break a silence - silence is worse for the customer
    * listen and key off of the customer (not hwat you want to hear)
    * do not disparage or degrade competition
    * plan each sales call - have objectives for each call - e.g. first call - goal is to get a 2nd one
    * do not waste time after you got your objective
    * sale is about people and relationships

## Measuring effectiveness
* summary measures for whcih sales and marketing are responsible for - e.g. sales volume, market share, profitability

### Behavior vs Performance
* behavior is what ppl do
* performance is contributions to the end goal

### Efficiency vs Effectiveness
* efficiency refers to “doing things right” while effectiveness refers to “doing the right things.”
* efficiency - cost efficiency
* effectiveness - 
* most common metric is Expense/Revenue ratio

### Individual vs Aggregate measures
* e.g. num sales calls, quality and length of calls, amount of service, amount of selling

### Short vs Long term effectiveness
* short term - quotas
* long term - indices of market share, profitability, customer satisfaction

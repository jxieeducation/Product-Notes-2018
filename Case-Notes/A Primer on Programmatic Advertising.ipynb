### Programmatic Advertising Ecosystem
* AdCreation 
    * reach potential customers in a cost-effective way to influence or create purchase
    * work with a media agency to create a strategy, campaign, and the ad
* Customer Targeting 
    * Data Management Platforms (DMP) purchase information from data suppliers
    * warehouse for housing, packaging and selling information
* placement
    * ad agencies place ads using its agency trading desk (ATD) and ad server
    * DSP, online platform where advertisers purchase inventory on connected ad exchanges and networks
    * DSP centralizes bidding and reporting to a single interface
    * ad exchange vs ad network - network has fixed CPMs, 
    * vertical networks are fully transparent on inventory purchased and which site its displayed
    * horizontal networks have more affordable pricing, but no full transparency, typically prepurchased inventory and targeted sites from ad exchange
    * publishers use a SSP to manage and sell their inventory, set price and audience type
    * programmatic channels
        * private marketplace - advertisers know precisely where their ad is placed
        * preferred deals - advertiser agrees to pay a fixed price, publisher gives advertiser the first right of refusal, publisher gets more control on quality of ad
        * programmatic direct - one to one between publisher and advertiser
